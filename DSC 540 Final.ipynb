This is a script to obtain data points for movies in HTML format from the IMDB website. Our goal is to get 5 data points for a subset of movies: Name of the movie; Year of release; IMDB rating; Metascore; and Number of votes.

```python
from requests import get

moviesurl = 'http://www.imdb.com/search/title?release_date=2017&sort=num_votes,desc&page=1'

response = get(moviesurl)

print(response.text[:500])
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')

type(soup)
```

```python
#use findall method to select the desired div containers
movie_containers = soup.find_all('div', class_ = 'lister-item mode-advanced')

print(type(movie_containers))

print(len(movie_containers))
```

```python
names = []
years = []
imdbrate = []
metascores = []
votes = []


for container in movie_containers:
    #remove movies who do not have a value for metascore
    if container.find('div', class_ = 'ratings-metascore') is not None:
        name = container.h3.a.text
        names.append(name)
        
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
        
        imdb = float(container.strong.text)
        imdbrate.append(imdb)
        
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
        
        vote = container.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
```

```python
import pandas as pd

test_df = pd.DataFrame({'movie': names,
'year': years,
'imdb': imdbrate,
'metascore': metascores,
'votes': votes
})

print(test_df.info())

test_df
```
